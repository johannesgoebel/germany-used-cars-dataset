```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df.info()
```

```python
df.isna().sum()
```

```python
df["brand"].unique()
```

```python
numeric_columns = df.select_dtypes(include=['number'])  # Select only numeric columns
```

```python
brand_avg = df.groupby('brand')[numeric_columns.columns].mean().round(2)

brand_avg
```

```python
brand_sum = df.groupby('brand')[numeric_columns.columns].sum()

brand_sum
```

```python
import csv

brand_sum.to_csv('sum_star_data.csv' , quoting=csv.QUOTE_ALL)
brand_avg.to_csv("avg_star_data.csv", quoting=csv.QUOTE_ALL)
```
